# Question 1 : What is the difference between multithreading and multiprocessing?

# Ans.

| **Aspect**        | **Multithreading**                                                                                                 | **Multiprocessing**                                                                                                     |
| ----------------- | ------------------------------------------------------------------------------------------------------------------ | ----------------------------------------------------------------------------------------------------------------------- |
| **Definition**    | Multithreading means running multiple threads (smaller units of a process) simultaneously within a single process. | Multiprocessing means running multiple processes simultaneously, each with its own Python interpreter and memory space. |
| **Memory Usage**  | Threads share the same memory space of the parent process.                                                         | Each process has its own independent memory space.                                                                      |
| **Communication** | Easier and faster (since memory is shared).                                                                        | Slower and more complex (uses inter-process communication).                                                             |
| **Crash Impact**  | If one thread crashes, it can affect the whole process.                                                            | If one process crashes, other                                                                                           |


# Question 2: What are the challenges associated with memory management in Python?

# Ans.
Python handles memory management automatically using its built-in garbage collector and dynamic memory allocation, but there are still several challenges:
| **Challenge**                        | **Explanation**                                                                                                                                                       |
| ------------------------------------ | --------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1. Memory Leaks**                  | Occur when objects are no longer needed but are still referenced, preventing the garbage collector from freeing the memory.                                           |
| **2. Reference Cycles**              | When two or more objects reference each other, making them unreachable but still existing in memory. The garbage collector may not immediately detect and clean them. |
| **3. High Memory Usage**             | Python objects have extra overhead (metadata), so programs using many small objects can consume more memory compared to lower-level languages like C.                 |
| **4. Global Interpreter Lock (GIL)** | The GIL prevents true parallel execution of threads, which can lead to inefficient CPU and memory usage in multithreaded programs.                                    |
| **5. Fragmentation**                 | Continuous allocation and deallocation of objects can fragment memory, reducing performance and available space.                                                      |
| **6. Manual Resource Management**    | Some resources (like file handles or database connections) must still be manually released using context managers (`with` statements) to avoid memory issues.         |



```python
'''Question 3:Write a Python program that logs an error message to a log file when a
division by zero exception occurs.'''
# Ans.
import logging

# Configure logging to write messages to a file
logging.basicConfig(filename='error.log',
                    level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

try:
    # Example: division operation
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)

except ZeroDivisionError:
    logging.error("Division by zero error occurred.")
    print("Error: Cannot divide by zero. Check error.log for details.")
```

```python
'''Question 4:Write a Python program that reads from one file and writes its content to
another file.
'''
#Ans.
# Open the source file in read mode and destination file in write mode
with open("source.txt", "r") as src_file:
    content = src_file.read()   # Read all contents from the source file

with open("destination.txt", "w") as dest_file:
    dest_file.write(content)    # Write the content to the destination file

print("File content copied successfully!")
```

#Question 6: What are the differences between NumPy arrays and Python lists?
#Ans.

| **Feature**            | **Python List**                                                                  | **NumPy Array**                                                                      |
| ---------------------- | -------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------ |
| **Type of elements**   | Can hold elements of **different data types** in the same list.                  | Requires **all elements to be of the same data type**.                               |
| **Memory efficiency**  | Less memory efficient (stores extra information for each element).               | More memory efficient due to **homogeneous, contiguous storage**.                    |
| **Performance**        | Slower for large-scale numerical computations.                                   | Faster, because operations are **vectorized** and implemented in C.                  |
| **Operations**         | Operations (like addition, multiplication) are **element-wise only with loops**. | Supports **vectorized operations** (e.g., `array1 + array2`) without explicit loops. |
| **Functionality**      | Limited built-in mathematical and statistical operations.                        | Offers **powerful mathematical, statistical, and linear algebra functions**.         |
| **Resizing**           | Dynamic — can easily append or remove elements.                                  | Fixed size — resizing creates a **new array**, not in-place.                         |
| **Slicing & Indexing** | Supports basic slicing and indexing.                                             | Supports **advanced slicing, boolean indexing, and multi-dimensional arrays**.       |


#Question 7:Explain the difference between apply() and map() in Pandas.
#Ans.
| **Feature**          | **`map()`**                                   | **`apply()`**                                                                                       |
| -------------------- | --------------------------------------------- | --------------------------------------------------------------------------------------------------- |
| **Used on**          | Primarily **Series**.                         | Can be used on **Series and DataFrame**.                                                            |
| **Function type**    | Works element-wise on each value of a Series. | Can apply a function **element-wise** (Series) or **row/column-wise** (DataFrame).                  |
| **Return type**      | Returns a **Series** of the same length.      | Returns a **Series or DataFrame** depending on context and function.                                |
| **Flexibility**      | Limited to **element-wise transformations**.  | More flexible — can apply **custom functions**, **aggregate**, or **transform** along rows/columns. |
| **Example use case** | Convert all strings in a Series to uppercase. | Sum rows/columns in a DataFrame or apply a complex custom function.                                 |

```python
'''Question 10: Calculate the correlation matrix using Seaborn and visualize it with a
heatmap.
'''
#Ans.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 65000, 70000, 80000],
    'Experience': [2, 5, 7, 10, 12]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Print correlation matrix
print(corr_matrix)

# Visualize correlation matrix as a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
